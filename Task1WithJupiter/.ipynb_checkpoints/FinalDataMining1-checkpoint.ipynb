```python
def groups_per_friend(id):    # Поиск  типов групп  по айдишнику
    try:
        groups = api.groups.get(user_id =id,extended = True,fields='activity')
        activitys=[]
        for i in groups['items']:
            activitys.append(i['activity'])
        
        return activitys
    except:
        print("private")
```

```python
act1 = groups_per_friend(id)
print(act1)
act2=[]
for i in act1:
    if(not (i in act2)):  ## мои типы групп (убрал дупликаты , но это не нужно, потом всё равнл перевдим в set)
        act2.append(i)
print(act2)
```

```python
friends = friends_get()
friendsId =friends["items"]   # Получение айдишников моих друзей
#print(friendsId)
def trueIdsearcher():
    for i in friendsId:
        try:
            temp =groups_per_friend(int (i))
            if(len(list(set(temp) & set(act2)))!=0):    # Если пересечение моего массива типов групп с массивом типов групп
                friend = api.users.get(user_id=int(i))  # друзей ненулевое , тогда вывожу друга 
                print(friend)
        except:
            print("Аккоунт приватный")   # доступ к акку закрыт , можно вообще ничего не выводить
            pass
            
trueIdsearcher()            
    
    
```
